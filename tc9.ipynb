```python
df=pd.DataFrame({'sname':['A','B','C','D','E'],
                 'kan':[8,78,100,76,990],
                 'Eng':[17,98,23,43,78],
                 'Hin':[10,90,98,78,988],
                 'Sci':[7,68,97,32,67],
                 'SS':[10,98,97,67,145],
                 'Mat':[7,56,10,79,90]})
df
```

```python
df['per']=(df['kan']+df['Eng']+df['Hin']+df['Sci']+df['SS']+df['Mat'])/6
df
```

```python
res=[]
for i in df['per']:
  if i>100 or i<0:
    res.append('invalid')
  elif i>=85:
    res.append('FCD')
  elif i>=60:
    res.append('FC')
  elif i>=50:
    res.append('SC')
  elif i>=35:
    res.append('JP')
  else:
    res.append('Fail')
df['result']=res
df
```
